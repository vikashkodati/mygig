```python
import torch
from transformers import pipeline
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import twitchio
```

```python
# Load the sentiment analysis pipeline using the default BERT model
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
# Test the model with a sample sentence
sample_text = "I love this stream!"
result = sentiment_pipeline(sample_text)

# Display the result
print(result)  # Should output the sentiment label and score
```

```python
chat_messages = [
    "This is an amazing stream!",
    "I don't like this content.",
    "The chat is so lively and fun!",
    "This is boring."
]

# Analyze sentiment for each message
for message in chat_messages:
    sentiment = sentiment_pipeline(message)
    print(f"Message: '{message}' | Sentiment: {sentiment}")
```
